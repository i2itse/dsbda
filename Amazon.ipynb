```python
import pandas as pd
import numpy as np

df = pd.read_csv("Amazonbookreviews.csv")

# Create row-wise subsets
subset1 = df.iloc[:100]
subset2 = df.iloc[100:200]

print("Subset 1:")
print(subset1.head())

print("\nSubset 2:")
print(subset2.head())
```

```python
# Merge subsets row-wise
merged_df = pd.concat([subset1, subset2])

print("Merged Data:")
print(merged_df.head())
```

```python
# Sort by review score (common column)
df_sorted = df.sort_values(by='overallRating', ascending=False)

print("Sorted Data:")
print(df_sorted.head())
```

```python
df_transposed = df.T

print("Transposed Data:")
print(df_transposed.head())
```
