```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PowerTransformer
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
data = pd.read_csv('AirQuality.csv',delimiter=";")
data
```

```python
data.drop(columns=['Unnamed: 15','Unnamed: 16','Date','Time'],inplace=True)
```

```python
data
```

```python
cols_to_fix = ['CO(GT)','C6H6(GT)','T','RH','AH']
for col in cols_to_fix:
  data[col] =data[col].str.replace(',','.')
  data[col]= data[col].astype(float)
```

```python
data
```

```python
data=data.astype({
    'CO(GT)':'float64','C6H6(GT)':'float64','T':'float64','RH':'float64','AH':'float64'
})
```

```python
data.dropna(inplace=True)
```

```python
data.drop_duplicates(inplace=True,ignore_index=True)
```

```python
#DATA TRANSFORMATION
```

```python
def remove_outliers(data, column_name, lower_quantile=0.01, upper_quantile=0.99):
  lower_bound = data[column_name].quantile(lower_quantile)
  upper_bound = data[column_name].quantile(upper_quantile)
  return data[(data[column_name] >= lower_bound) & (data[column_name] <= upper_bound)]

df=data.copy()
for col in df.columns:
    ds=remove_outliers(df,col)
```

```python
df = df.reset_index(drop=True)
```

```python
pt = PowerTransformer(method = 'yeo-johnson')
df=pd.DataFrame(pt.fit_transform(df),columns=df.columns)
```

4) ERROR CORRECTION

```python
print('missing values: \n',df.isnull().sum())
```

```python
z_scores=(ds-ds.mean())/ds.std()
extreme_values=(np.abs(z_scores)>3).sum()

print("\n Extreme values(>3 std dev:P\n",extreme_values)
```

```python
#model Building
x = df.drop('AH',axis=1)
y = df['AH']

x_train , x_test, y_train, y_test = train_test_split(x,y, train_size=0.8,random_state=6, shuffle=True )
```

```python
lr_model = LinearRegression()
lr_model.fit(x_train, y_train)
lr_pred = lr_model.predict(x_test)
```

```python
dt_model = DecisionTreeRegressor()
dt_model.fit(x_train, y_train)
dt_pred = dt_model.predict(x_test)
```

```python
def evaluate_model(name, y_true, y_pred):
    r2 = r2_score(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    print(f"\n{name} Model:")
    print(f" R² Score     : {r2:.4f}")
    print(f" MAE          : {mae:.4f}")
    print(f" MSE          : {mse:.4f}")
    print(f" Accuracy (%) : {r2*100:.2f}%")
```

```python
evaluate_model("Linear Regression", y_test, lr_pred)
evaluate_model("Decision Tree Regressor", y_test, dt_pred)
```
