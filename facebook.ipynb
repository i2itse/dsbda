```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset_Facebook.csv',sep=";")
```

```python
df.head(10)
```

```python
#create subset 1 
subset1 = df[["Type","Category","Post Month"]]
print(subset1.head(10))
```

```python
#create subset 2 
subset2 = df[["comment","Type","like"]]
print(subset2.head(10))
```

```python
#merge 2 subsets 
merged = pd.concat([subset1,subset2] , axis=1)
print(merged.head(10))
```

```python
#sort values
merged = merged.sort_values(by=["comment"],ascending=False)
print(merged.head(10))
```
